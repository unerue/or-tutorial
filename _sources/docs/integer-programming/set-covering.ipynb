## 최소의 파출소 후보지역을 구하기 위한 0-1정수계획모형

$$\begin{align*}
  & \text{minimize }   &   Z=7&X_{1} + 8X_{2} + 3X_{3}       \\
  & \text{subject to } & \,  3&X_{1} + 4X_{2} + 2X_{3} \le 6 \\
  & \text{and}         & \,   &X_{j} = 0 \text{ or } 1 (j=1,2,3)
\end{align*}$$
